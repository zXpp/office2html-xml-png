```python
d="/media/zzpp220/Data/Linux_Documents/DOWNLOAD/python-DataAnalysis/wordcloud"

text=open(path.join(d,'translator.txt')).read().decode('gbk')

#text##中文在linux内部的形式是Unicode，就是下面这些
```
